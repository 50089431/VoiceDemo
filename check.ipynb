```python
import os  
import base64
from openai import AzureOpenAI  
from azure.identity import DefaultAzureCredential, get_bearer_token_provider
```

```python
       
endpoint = "https://demoopenairesource.openai.azure.com/"
deployment = "gpt-4"
      
# Initialize Azure OpenAI Service client with Entra ID authentication
token_provider = get_bearer_token_provider(  
    DefaultAzureCredential(),  
    "https://cognitiveservices.azure.com/.default"  
)  
  
client = AzureOpenAI(  
    azure_endpoint=endpoint,  
    azure_ad_token_provider=token_provider,  
    api_version="2024-05-01-preview",  
)  
  

# IMAGE_PATH = "YOUR_IMAGE_PATH"
# encoded_image = base64.b64encode(open(IMAGE_PATH, 'rb').read()).decode('ascii')
chat_prompt = [
    {
        "role": "system",
        "content": [
            {
                "type": "text",
                "text": "Answer the question concisely while providing a brief explanation or additional context if needed.\n\n# Output Format\n\n- A single sentence stating the capital of France.\n- Optionally, include any relevant additional context (e.g., cultural or historical significance) after the direct answer.\n\n# Example\n\nInput: \"What is the capital of Italy?\"  \nOutput: \"The capital of Italy is Rome, which is renowned for its historical landmarks such as the Colosseum and Vatican City.\""
            }
        ]
    }
] 
    
# Include speech result if speech is enabled  
messages = chat_prompt 
completion = client.chat.completions.create(  
    model=deployment,  
    messages=messages,
    max_tokens=800,  
    temperature=0.7,  
    top_p=0.95,  
    frequency_penalty=0,  
    presence_penalty=0,
    stop=None,  
    stream=False  
)  
  
print(completion.to_json())  
```
